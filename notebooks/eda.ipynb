```python
%pip install pandas matplotlib ipykernel
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('../data/raw/collected.csv')
df['text_len'] = df['text'].fillna('').astype(str).str.len()
df['label'].value_counts().plot(kind='bar', title='Class distribution')
plt.show()
```

```python
df['text_len'].hist(bins=30)
plt.title('Text length distribution')
plt.show()
```
