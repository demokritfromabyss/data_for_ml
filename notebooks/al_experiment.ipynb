# Active Learning Experiment


```python
import pandas as pd
from agents.al_agent import ActiveLearningAgent


df = pd.read_csv('../data/raw/clean.csv')
agent = ActiveLearningAgent()
initial_df, pool_df, test_df = agent.prepare_splits(df, initial_size=80)
entropy_history = agent.run_cycle(initial_df, pool_df, strategy='entropy', n_iterations=5, batch_size=25, test_df=test_df)
random_history = agent.run_cycle(initial_df, pool_df, strategy='random', n_iterations=5, batch_size=25, test_df=test_df)
pd.DataFrame(entropy_history), pd.DataFrame(random_history)
```
